```python
from selenium import webdriver
import pandas as pd

url = "http://checker.sindheducation.gov.pk/CheckBiometrics.aspx"
# create a new Firefox session
driver = webdriver.Firefox(executable_path='E:/geckodriver-v0.29.0-win64/geckodriver.exe')
driver.implicitly_wait(30)
driver.get(url)
```

```python
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup
import re
import os

# read data_file
data_f = pd.read_csv('E:/DGME/cnic.csv')
data_f = data_f[1:4]
#data_f # 393025
```

```python
from selenium.webdriver.support.ui import Select
import time
import re
import csv 
# Search DDL
search_ddl = driver.find_element_by_id("ContentPlaceHolder1_ddlSearchBy")
select = Select(search_ddl)
select.select_by_index(1)
time.sleep(5)

for index, row in data_f.iterrows():
    search_text = driver.find_element_by_id("ContentPlaceHolder1_txtSearch")
    search_text.click()
    search_text.clear()
    search_text.send_keys(row['CNIC'])
    search_text.send_keys(Keys.ENTER)
    time.sleep(1)   
    search_button = driver.find_element_by_id("ContentPlaceHolder1_btn_search")
    search_button.click()
    time.sleep(5)   
    #Create a dataframe to store all of the scraped table data
    df = pd.DataFrame(columns = ["Reg ID", "Emp. Name", "Fathers Name", "P.ID", "Gender", "Designation", "BPS", "Type", "SEMIS ID (DC)", "Inst. Name (DC)", "District (DC)", "SEMIS ID (M&E)", "Inst. Name (M&E)", "District (M&E)", "Status" ] )
    for i in range(3):
        try:
            mytable = driver.find_element_by_css_selector("table[id='ContentPlaceHolder1_gvDCDetails']")
            # Read in all of the data into the dataframe
            for row in mytable.find_elements_by_css_selector('tr'):
                row_list = []
                #Add to dataframe accordingly
                for cell in row.find_elements_by_css_selector('td'):
                    cell_reading = cell.text
                    row_list.append(cell_reading)
                    print(row_list)
                #Add the list as a row, if possible 
                try:
                    a_series = pd.Series(row_list, index = df.columns)
                    df = df.append(a_series, ignore_index=True)
                except:
                    print("Could not append: " + str(row_list))
            break
        except:
            driver.implicitly_wait(5)
#Wait 10 seconds for the load
    driver.implicitly_wait(10)

```
