```python
def autolabel(rects):
    """Attach a text label above each bar , displaying its height."""
    for rect in rects:
        height = rect.get_height()
        #plt.text(rect.get_x() + rect.get_width()/2.0, height, '%d' % int(height), ha='center', va='bottom') 
        plt.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 7),  # 4 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom', size = 12, color='white')
```

```python
import seaborn as sns
#plt.figure(figsize=(12, 5))

sns.set(rc={'figure.figsize':(11.7,8.27)})
ax = plt.gca()
ax.set_facecolor('xkcd:black')

bar = sns.barplot(x="Food", y="Calories", data=df, palette="Blues_d")
for p in bar.patches:
    bar.annotate(format(p.get_height()), (p.get_x() + p.get_width() / 2., p.get_height()), ha = 'center', va = 'center', color = 'white', xytext = (0, 10), textcoords = 'offset points')
plt.xticks(style='italic', size=12)
plt.ylabel('NUMBER Of CALORIES', size=12, style='italic')
plt.xlabel('')
plt.title('Calories Per 100g of different foods', style = 'italic')

def change_width(ax, new_value) :
    for patch in bar.patches :
        current_width = patch.get_width()
        diff = current_width - new_value

        # we change the bar width
        patch.set_width(new_value)

        # we recenter the bar
        patch.set_x(patch.get_x() + diff * .5)

change_width(bar, .35)

plt.show()
```

```python
def autolabel(rects):
    """Attach a text label above each bar , displaying its height."""
    for rect in rects:
        height = rect.get_height()
        #plt.text(rect.get_x() + rect.get_width()/2.0, height, '%d' % int(height), ha='center', va='bottom') 
        plt.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 7),  # 4 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom', size = 15, color='black')
```

```python
def autolabel_line(xs, ys):
    for x,y in zip(xs,ys):
        label = "{:.2f}".format(y)
        plt.annotate(label, # this is the text
                 (x,y), # this is the point to label,    
                 textcoords="offset points", # how to position the text
                 xytext=(0,7), # distance from text to points (x,y)
                 ha='center') # horizontal alignment can be left, right or center
```

```python
def change_width(ax, new_value, bar) :
    for patch in bar.patches :
        current_width = patch.get_width()
        diff = current_width - new_value

        # we change the bar width
        patch.set_width(new_value)

        # we recenter the bar
        patch.set_x(patch.get_x() + diff * .5)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

file2 = pd.read_excel('E:/dashboards/sdg3/Pakistan at glance.xlsx', sheet_name='immunization pct_change')
#df2 = pd.DataFrame(file2)
df2= pd.DataFrame(file2).reset_index(drop='True')
df2['progress_change'] = df2['Progress'].diff()
df2['pct_change'] = df2['Progress'].pct_change()*100
df2.fillna(0)


fig, ax2 = plt2.subplots(figsize=(20, 10))
color = 'tab:red'
ax2.set_ylabel('Change in Progress (%)', color='black' , size=15)  # we already handled the x-label with ax1
bar2 = ax2.plot(df2['Year'], df2['pct_change'], color=color, marker='o')
ax2.tick_params(axis='y', labelcolor=color)
ax2.set_title('Child Immunization', color='green', size=15)

autolabel_line(df2['Year'], df2['pct_change'].round())

plt.grid(axis='y', which = 'major', linestyle='--')
```

```python
def pct_change_inc(data):
    d = []
    for i in range(len(data)):
        if i == 0:
            d.append(data[0]/data[0])
        else:
            d.append(((data[i]-data[i-1])/data[i-1] * 100).round(1))
    return d 
    
def pct_change_dec(data):
    d = []
    for i in range(len(data)):
        if i == 0:
            d.append(0)
        else:
            d.append(((data[i-1]-data[i])/data[i-1] * 100).round(1))
    return d 
    
    
```
