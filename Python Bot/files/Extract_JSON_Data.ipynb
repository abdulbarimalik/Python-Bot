```python
import json
import os
import csv
import pandas as pd

INPUT_PATH ="E:/json_files/"
column_name =  ["Name", "Email", "City","messageCount","Messages"]
dict = {}
list = []
index = 0
for file in os.listdir(INPUT_PATH):
    if file.endswith('.json') or file.endswith('.JSON'):
        INPUT_FILE = os.path.join(INPUT_PATH, file)
        with open(INPUT_FILE) as f:
            json_data = json.load(f)
            #print(json_data)
            dict[column_name[0]] = json_data["visitor"]["name"]
            dict[column_name[1]] = json_data["visitor"]["email"]
            dict[column_name[2]] = json_data["location"]["city"]
            dict[column_name[3]] = json_data["messageCount"]
    #        print(len(json_data["messages"]))
            loop = len(json_data["messages"])
            print(loop)
            for i in range(1, loop):
                list.insert(i, json_data["messages"][i]["msg"]) 
                dict[column_name[4]] = list
            print(dict)
#             print(dict.keys())
#             print(dict.values())     
        with open(r'E:/Recordset.csv', 'a', newline='') as f_object:
            dictwriter_object = csv.DictWriter(f_object, fieldnames=column_name)            
            if index == 0:
                dictwriter_object.writeheader()      
            dictwriter_object.writerow(dict)
            index = index + 1
            dict.clear()
            list.clear()
```
